```python
from ecpli.benchmark1d.CutoffBenchmarkDataset1D import CutoffBenchmarkDataset1D
import pickle
import ecpli
import matplotlib.pyplot as plt
import numpy as np
from ecpli.plot_helper import plot_profile_likelihood, plot_ts
```

```python
infile = "test/benchmark_NI0D0CPRHM_limit_UnconstrainedLR_ConstrainedLR.pickle"
with open(infile, "rb") as fin:
    dataset = pickle.load(fin)
```

```python
for limit in dataset["limit_list"]:
    lambda_ul = limit[1]
    lambda_true = dataset["dataset_parameter"]["true_parameter"]["lambda_true"].value
    limit_method = limit[0]
    
    if isinstance(limit_method, ecpli.LRBase):
        plt.figure(figsize=(16, 9))
        plot_profile_likelihood(plt,
                                limit_method,
                                lambda_true=lambda_true)
        
        plt.figure(figsize=(16, 9))
        plot_ts(plt,
                limit_method,
                lambda_true=lambda_true,
                lambda_ul=lambda_ul)
```
